```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./traffic_sao_paulo.csv',sep=';')
```

```python
df['Slowness in traffic (%)'] = df['Slowness in traffic (%)'].str.replace(',','.')
df['Slowness in traffic (%)'] = df['Slowness in traffic (%)'].astype(float)
```

```python
days = ['monday','tuesday','wednesday','thursday','friday']
traffic_per_day = {}
```

```python
df['Hour (Coded)'].unique()
```

```python
for i,day in zip(range(0,df.shape[0],27),days):
    traffic_per_day[day] = df[i:i+27]
```

```python
for day in traffic_per_day.keys():
    traffic_per_day[day].plot.line(x='Hour (Coded)',y='Slowness in traffic (%)')
    plt.title(f'{day} traffic')
    plt.ylim([0,25])
    plt.show()
```

```python
traffic_per_day['monday'].plot.line(
    x='Hour (Coded)',
    y='Slowness in traffic (%)'
    )
plt.title('Hour (Coded) vs Slowness in traffic (%)')
plt.ylim([0,25])
plt.show()
```
